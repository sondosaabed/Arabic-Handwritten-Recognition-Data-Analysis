```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import itertools
```

```python
df = pd.read_csv("./KhattTech UX Research (Responses) - Form responses 1.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.columns = [col.rstrip() for col in df.columns]
```

```python
# Rename columns for clarity
df.rename(columns={
    "How frequently do you need to extract Arabic handwritten text from images?": "Usage Frequency",
    "What is your primary use case for extracting Arabic handwritten text?": "Primary Use Case",
    "What is your biggest challenge when trying to extract Arabic handwritten text?": "Biggest Challenge",
    "How satisfied are you with the accuracy of existing Arabic Handwritten Recognition tools?": "Satisfaction",
    "What features would you like in an Arabic Handwritten Recognition app?": "Desired Features",
    "How important is it for you to share extracted text through messaging apps (e.g., WhatsApp, Signal, Messenger)?": "Importance of Sharing",
    "Would you use a mobile app for Arabic Handwritten Recognition? (Yes/No)\nIf No, why not?": "Would Use Mobile App",
    "Do you have any additional feedback or suggestions for improving Arabic Handwritten Recognition technology?": "Additional Feedback",
    "If a high-accuracy Arabic Handwritten Recognition app were available, how much would you be willing to pay for it?": "Willingness to Pay",
    "Which of the following would be dealbreakers for you in an Arabic Handwritten Recognition app? (Select all that apply)": "Dealbreakers"
}, inplace=True)
```

```python
df.info()
```

```python

# Handle multiple-choice responses by splitting into lists
def split_multiple_choices(column):
    return df[column].apply(lambda x: x.split(", ") if isinstance(x, str) else [])
```

```python
df["Desired Features"] = split_multiple_choices("Desired Features")
df["Dealbreakers"] = split_multiple_choices("Dealbreakers")
df["Biggest Challenge"] = split_multiple_choices("Biggest Challenge")
```

```python
df.head()
```

```python
# Get value counts
freq_counts = df["Usage Frequency"].value_counts()
colors = sns.color_palette('pastel')[0:len(freq_counts)]

# Plot
plt.figure(figsize=(8, 8))
plt.pie(freq_counts, labels=freq_counts.index, colors=colors, autopct='%1.1f%%', startangle=140, wedgeprops={'edgecolor': 'black'})
plt.title('Frequency of Use', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# 2. Challenges in Extraction
print("\nTop Challenges:")
challenge_counts = df.explode("Biggest Challenge")["Biggest Challenge"].value_counts()
print(challenge_counts)
```

```python
challenge_series = pd.Series(challenge_counts).sort_values()

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=challenge_series.values, y=challenge_series.index, palette="pastel")

# Add value labels
for index, value in enumerate(challenge_series.values):
    plt.text(value + 0.2, index, str(value), va='center', fontweight='bold')

# Styling
plt.title('Top Challenges in Arabic Handwritten Text Extraction', fontsize=14, fontweight='bold')
plt.xlabel('Number of Mentions')
plt.ylabel('Challenges')
plt.xlim(0, max(challenge_series.values) + 2)
plt.tight_layout()
plt.show()
```

```python
# 3. User Satisfaction
print("\nUser Satisfaction:")
print(df["Satisfaction"].value_counts())
```

```python
satisfaction_counts = df["Satisfaction"].value_counts().sort_index()

# Calculate percentages
satisfaction_percent = (satisfaction_counts / satisfaction_counts.sum()) * 100

# Plot
plt.figure(figsize=(8, 6))
sns.barplot(x=satisfaction_counts.index, y=satisfaction_counts.values, palette="pastel")

# Add labels
for index, value in enumerate(satisfaction_counts.values):
    plt.text(index, value + 0.5, f"{value} ({satisfaction_percent.iloc[index]:.1f}%)", ha='center', fontweight='bold')

# Styling
plt.title('User Satisfaction with Arabic Handwritten Text Extraction Tools', fontsize=14, fontweight='bold')
plt.xlabel('Satisfaction Level (1 = Very Dissatisfied, 5 = Very Satisfied)')
plt.ylabel('Number of Responses')
plt.ylim(0, max(satisfaction_counts.values) + 3)
plt.tight_layout()
plt.show()
```

```python
# 4. Desired Features
print("\nDesired Features:")
desired_features_counts = df.explode("Desired Features")["Desired Features"].value_counts()
print(desired_features_counts)
```

```python
# Cleaned data
feature_counts = {
    "High-accuracy text recognition": 15,
    "Ability to edit extracted text before saving": 12,
    "Support for multiple handwriting styles": 11,
    "Downloading in multiple formats (PDF, TXT)": 11,
    "Integration with messaging & cloud apps": 5
}

# Convert to Series
feature_series = pd.Series(feature_counts).sort_values()

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=feature_series.values, y=feature_series.index, palette="pastel")

# Add value labels
for index, value in enumerate(feature_series.values):
    plt.text(value + 0.2, index, str(value), va='center', fontweight='bold')

# Styling
plt.title('Most Desired Features in Arabic Handwritten Text Extraction Tools', fontsize=14, fontweight='bold')
plt.xlabel('Number of Mentions')
plt.ylabel('Desired Features')
plt.xlim(0, max(feature_series.values) + 2)
plt.tight_layout()
plt.show()
```

```python
# 5. Market Demand
print(df["Would Use Mobile App"].value_counts())

print("\n")
print(df["Willingness to Pay"].value_counts())
```

```python
# Cleaned data
app_usage_counts = {
    "Yes": 14,
    "Sometimes": 1,
    "No / Prefer Web": 2
}

# Convert to Series
app_series = pd.Series(app_usage_counts)

# Pie chart
plt.figure(figsize=(7,7))
colors = sns.color_palette('pastel')[0:3]
plt.pie(app_series, labels=app_series.index, colors=colors, autopct='%1.1f%%', startangle=140, wedgeprops={'edgecolor':'black'})
plt.title('Willingness to Use a Mobile App for OCR', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Cleaned data
pay_counts = {
    "Free with ads": 9,
    "Wouldn’t pay": 4,
    "One-time fee": 3,
    "Monthly subscription": 1
}

pay_series = pd.Series(pay_counts).sort_values()

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=pay_series.values, y=pay_series.index, palette="pastel")

# Add value labels
for index, value in enumerate(pay_series.values):
    plt.text(value + 0.2, index, str(value), va='center', fontweight='bold')

plt.title('Willingness to Pay for Arabic OCR Services', fontsize=14, fontweight='bold')
plt.xlabel('Number of Responses')
plt.ylabel('Payment Preference')
plt.xlim(0, max(pay_series.values) + 2)
plt.tight_layout()
plt.show()
```

Do those who rate accuracy as poor show a higher willingness to pay?

```python
# Filter users who mention poor accuracy
df['Biggest Challenge'] = df['Biggest Challenge'].astype(str)  # Ensure text
accuracy_issue = df[df['Biggest Challenge'].str.contains('Poor accuracy', case=False)]

# Count their willingness to pay
plt.figure(figsize=(8, 5))
sns.countplot(data=accuracy_issue, x='Willingness to Pay', palette='Set2')
plt.title('Willingness to Pay for Users Reporting Poor Accuracy')
plt.xticks(rotation=30)
plt.show()

# Compare with general distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Willingness to Pay', palette='Set3')
plt.title('Overall Willingness to Pay Distribution')
plt.xticks(rotation=30)
plt.show()
```

- Are those who struggle with recognition accuracy more frequent users?  

```python
# Check usage frequency of those reporting poor accuracy
plt.figure(figsize=(8, 5))
sns.countplot(data=accuracy_issue, x='Usage Frequency', order=['Rarely', 'Monthly', 'Weekly'], palette='coolwarm')
plt.title('Usage Frequency of Users with Recognition Accuracy Issues')
plt.show()
```

- Is personal use associated with lower sharing needs?  

```python
# Filter personal use
personal_use = df[df['Primary Use Case'].str.contains('Personal note-taking', case=False)]

# Plot Importance of Sharing
plt.figure(figsize=(8, 5))
sns.histplot(personal_use['Importance of Sharing'], bins=4, kde=False, color='orange')
plt.title('Importance of Sharing - Personal Use')
plt.xlabel('Importance of Sharing Score')
plt.show()

# Compare with others
others = df[~df['Primary Use Case'].str.contains('Personal note-taking', case=False)]

plt.figure(figsize=(8, 5))
sns.kdeplot(personal_use['Importance of Sharing'], label='Personal Use', fill=True)
sns.kdeplot(others['Importance of Sharing'], label='Other Uses', fill=True, color='green')
plt.title('Importance of Sharing: Personal vs. Other Uses')
plt.legend()
plt.show()
```

- Do users with high privacy concerns hesitate to pay?  


```python
# Filter rows where 'Privacy concerns' is in the Dealbreakers list
privacy_concern = df[df['Dealbreakers'].apply(lambda x: 'Privacy concerns' in x)]

# Visualize their willingness to pay
plt.figure(figsize=(8, 5))
sns.countplot(data=privacy_concern, x='Willingness to Pay', palette='magma')
plt.title('Willingness to Pay - Users with Privacy Concerns')
plt.xticks(rotation=30)
plt.show()
```

- Are those who demand high accuracy also unwilling to tolerate a high price?  


```python
# Filter users who desire high accuracy text recognition
high_accuracy = df[df['Desired Features'].apply(lambda x: 'High-accuracy text recognition' in x)]

# Plot their willingness to pay
plt.figure(figsize=(8, 5))
sns.countplot(data=high_accuracy, x='Willingness to Pay', palette='BuGn')
plt.title('Willingness to Pay - High Accuracy Demanders')
plt.xticks(rotation=30)
plt.show()
```
